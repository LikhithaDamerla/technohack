```python
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].apply(lambda date:date.month)
df['year'] = df['date'].apply(lambda date:date.year)

#data visualization house price vs months and years
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
a=df.groupby('month').mean()['price']
plt.ylabel('Price',fontsize=20)
plt.xlabel('Month',fontsize=20)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
a.plot()

plt.subplot(1,2,2)
b=df.groupby('year').mean()['price']
plt.ylabel('Price',fontsize=20)
plt.xlabel('Year',fontsize=20)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
b.plot()
```

```python
df = df.drop('date',axis=1)
df = df.drop('id',axis=1)
df = df.drop('zipcode',axis=1)
```

```python
X = df.drop('price',axis =1).values
y = df['price'].values
```

```python
print(X.shape)
X
```

```python
print(y.shape)
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python

from sklearn.preprocessing import StandardScaler
s_scaler = StandardScaler()
X_train = s_scaler.fit_transform(X_train.astype(np.float))
X_test = s_scaler.transform(X_test.astype(np.float))
```

```python
print(np.max(X_train))
print(np.min(X_train))
```

```python
print(np.max(X_test))
print(np.min(X_test))
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()  
regressor.fit(X_train, y_train)
```

```python
print(regressor.intercept_)
print(regressor.coef_)
```

```python
y_pred = regressor.predict(X_test)
```

```python
coeff_df = pd.DataFrame(regressor.coef_, df.drop('price',axis =1).columns, columns=['Coefficient']) 
coeff_df
```

```python
plt.figure(figsize=(10,5))
residuals = (y_test- y_pred)
sns.histplot(residuals, color='green')
```

```python
y_pred = regressor.predict(X_test)
dff = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
dff.head(10)
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, y_pred))  
print('MSE:', metrics.mean_squared_error(y_test, y_pred))  
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('VarScore:',metrics.explained_variance_score(y_test,y_pred))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))
model.compile(optimizer='Adam',loss="mean_squared_error")
```

```python
model.fit(x=X_train,y=y_train,
          validation_data=(X_test,y_test),
          batch_size=128,epochs=400)
model.summary()
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot(figsize=(12,8))
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))  
print('MSE:', metrics.mean_squared_error(y_test, y_pred))  
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('VarScore:',metrics.explained_variance_score(y_test,y_pred))
```
